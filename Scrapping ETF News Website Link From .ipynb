```python
import json
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
headers = { "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}
```

```python
response = requests.get("https://www.google.com/search?q=QQQ ETF&gl=us&tbm=nws&num=300", headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
news_results = []
```

```python
for el in soup.select("div.SoaBEf"): news_results.append({
           "link": el.find("a")["href"],
           "title": el.select_one("div.MBeuO").get_text(),
           "snippet": el.select_one(".GI74Re").get_text(),
            "date": el.select_one(".LfVVr").get_text(),
            "timestamp": datetime.datetime.now().isoformat(),
             "source": el.select_one(".NUnG9d span").get_text()
            }
        )
 
print(json.dumps(news_results, indent=2))
```

```python
i =1
while i<=98:
    news_results[i]['link']
    print(news_results[i]['link'])
    i +=1
```

```python
import pandas as pd

pd.DataFrame({'link': [news_results[i]['link']]})
```

```python
news_website_link = pd.DataFrame()
i =1
while i<=100:
    news_results[i]['link']
    print(news_results[i]['link'])
    news_website_link = pd.concat([news_website_link, pd.DataFrame({'link': [news_results[i]['link']]})], ignore_index=True)
    i +=1
```
